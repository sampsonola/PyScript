```python
df_long = pd.melt(df, id_vars=['Country'], var_name='Year', value_name='Value')
# Convert 'Year' to datetime
df_long['Year'] = pd.to_datetime(df_long['Year'], format='%Y')
```

```python
df_long.head()
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Filter data for a single country for simplicity
country_df = df_long[df_long['Country'] == 'Afghanistan'].set_index('Year')

# Fit ARIMA model
model = ARIMA(country_df['Value'], order=(1, 1, 1))
fit_model = model.fit()
print(fit_model.summary())
# Forecast future values
forecast = fit_model.forecast(steps=5)

# Plot the original data and forecast
plt.figure(figsize=(12, 6))
plt.plot(country_df.index, country_df['Value'], label='Original')
plt.plot(pd.date_range(country_df.index[-1], periods=6, freq='Y')[1:], forecast, label='Forecast', marker='o')
plt.title('ARIMA Forecast')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the time series
result = seasonal_decompose(country_df['Value'], model='additive', period=1)
result.plot()
plt.show()
```

```python
from prophet import Prophet

# Prepare data for Prophet
df_prophet = country_df.reset_index().rename(columns={'Year': 'ds', 'Value': 'y'})

# Initialize and fit the model
model = Prophet()
model.fit(df_prophet)

# Create a DataFrame to hold predictions
future = model.make_future_dataframe(periods=5, freq='Y')
forecast = model.predict(future)

# Plot the forecast
fig = model.plot(forecast)
plt.title('Facebook Prophet Forecast')
plt.xlabel('Year')
plt.ylabel('Value')
plt.show()
```
